### The sigmoid function


```python
def sigmoid(t):
    return 1.0/(1+exp(-t))

#visualize sigmoid function
x=arange(-5.0,5.0,0.1)
y=sigmoid(x)
plt.plot(x,y)
plt.show()
```

### Build the classifier
As is proved in the blog, to implement a logistic classifier, we do gradient ascend, which is simply to repeatly apply the following update to the weights:
$$w_{new}=w_{old}+\alpha (\sum_{i=1}^{m} (y_{i}-s(w^{T} x_{i}))x_{i})$$

Notice that $y_{i}-s(w^{T} x_{i})$ is a number, let's call it $E_i$. Suppose there are m training points. $x_{i}$ is a column vector, the j-th dimension of $w_{new}$ is the j-th dimension of $w_{old}$ plus the sum of product of $E_i$ and the j-th dimension of $x_i$, where $1\le i \le m $  

#### Load the data
Load the features and the corresponding labels. Notice that the features comes in rows.

```python
matrix=df.values
#get the features
f=matrix[:,0:2]
#create a column of ones
list_one=ones((f.shape[0],1))
#append this "1" column to the features
features=column_stack((f,list_one))
#get the lables
label_list=matrix[:,2]
labels=mat(label_list).transpose()
features[:10]
```

#### Calculate E_i

```python
dataSet=mat(features)
#number of training points
numberOfPoints=features.shape[0]

#number of features
numberOfFeatures=features.shape[1]

#initiate weight matrix with all 1s
weights=mat(ones((numberOfFeatures,1)))
weights

#wx, ech row i is the result for w^Tx_i
wx=sigmoid(dataSet*weights)

#calculate E_i for all the numberOfPoints training points. The resulting E_i comes in rows
E=labels-sigmoid(dataSet*weights)
#The resulting E is a column vector where row i correspond to E_i
E[:10]
```

#### Implement the update part of the weight

Observe E. Each row correspond to the error for each training points. Recall that the j-th dimension of $w_{new}$ is the j-th dimension of $w_{old}$ plus the sum of products of the j-dimension of $x_i$ and $E_i$ ($1\le i \le m$). So suppose the matrix of training data points is on the left, the number of columns of this matrix should equal to the number of training points and the number of rows should equal to the number of features

```python
dataSet.transpose()*E
```

The result is a column vector with 3 rows. This is what we want

#### Write the complete function
Now we have all the ingredients for writing a gradient ascent function for logistic regression. We can combine all the ingredients to write the function:

```python
#features: each row is a feature vector with 1 added
#classLabels: array of training labels
def gradeintAscent(features,classLabels):
    labels=mat(classLabels).transpose()
    dataSet=mat(features)
    numberOfPoints=features.shape[0]
    numberOfFeatures=features.shape[1]
    #a columns vector with weights
    weights=mat(ones((numberOfFeatures,1)))
    alpha=0.001
    maxCycles=500
    for k in range(maxCycles):
        error=labels-sigmoid(dataSet*weights)
        weights=weights+alpha*dataSet.transpose()*error
    return weights

weight=gradeintAscent(features,label_list)
weight
```

```python


cmap = {1: 'red', 0: 'blue'}
df.plot(x='x1', y='x2', kind='scatter', c=[cmap.get(c, 'black') for c in df.label])
x=arange(-5.0,5.0,0.1)
y=(-weight[2,0]-weight[0,0]*x)/weight[1,0]
plt.plot(x,y)
plt.show()
```
